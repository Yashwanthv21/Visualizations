```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('gdp.csv')
```

```python
df.head()
```

```python
cols = list(df.columns)
df = df[[cols[0], cols[2], cols[3], cols[4], cols[5], cols[6], cols[7]]]
```

```python
n_frames = 500

rows = []
for i in range(len(df)-1):
    r1 = df.iloc[i].tolist()
    r2 = df.iloc[i+1].tolist()
    year = r1[0]
    temp = []
    for j in range(1, len(r1)):
        diff = (r2[j] - r1[j])/n_frames
        ss = [r1[j] + idx*diff  for idx in range(n_frames)]
        temp.append(ss)
#         print(len(ss))
        
    temp = np.array(temp)
    temp = temp.T.tolist()
    
    for e in temp:
        e.insert(0, year)
        
    rows.extend(temp)
    rows.append(r2)
    
```

```python
dd = pd.DataFrame(rows, columns = df.columns)
dd.head()
```
